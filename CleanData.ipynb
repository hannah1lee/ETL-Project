```python
import pandas as pd
```

```python
# Cleaning U.S. Unemployment Data

us_data = "Resources/unemployment_data_us.csv"
us_df = pd.read_csv(us_data)
us_df.head()
```

```python
# Delete extra column and rows

del us_df['Date']
us_df.head()
```

```python
new_df = pd.DataFrame({"Primary_School" : us_df.groupby(["Year", "Month"])["Primary_School"].mean(),
                       "Date" : us_df.groupby(["Year", "Month"])["Date"].mean(),
                       "High_School" : us_df.groupby(["Year", "Month"])["High_School"].mean(),
                       "Associates_Degree" : us_df.groupby(["Year", "Month"])["Associates_Degree"].mean(),
                       "Professional_Degree": us_df.groupby(["Year", "Month"])["Professional_Degree"].mean(),
                       "White": us_df.groupby(["Year", "Month"])["White"].mean(),
                       "Black": us_df.groupby(["Year", "Month"])["Black"].mean(),
                       "Asian": us_df.groupby(["Year", "Month"])["Asian"].mean(),
                       "Hispanic": us_df.groupby(["Year", "Month"])["Hispanic"].mean(),
                       "Men": us_df.groupby(["Year", "Month"])["Men"].mean(),
                       "Women": us_df.groupby(["Year", "Month"])["Women"].mean()})
```

```python
clean_us_df = us_df.dropna()
clean_us_df.set_index(['Year', 'Month'])
```

```python
# Cleaning Ohio Unemployment Data

oh_data = "Resources/OhioUnemployment.xlsx"
oh_df = pd.read_excel(oh_data)
oh_df.head()
```

```python
oh_df.set_index(['Year', 'Month'])
```

```python
# Joining Data

merged = pd.merge(clean_us_df,oh_df, on=["Year",'Month'],how="inner")
merged.head()
```

```python
merged.to_csv('Resources/merged_unemployment.csv', index=False)
```
